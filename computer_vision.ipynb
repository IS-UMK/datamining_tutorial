O wiele naturalniejsz przetrzeni kolorw jest dla czowieka jest przestrze HSV

```python
from skimage import color

img = data.coffee()

#zwrc uwag, e teraz img jest typu zmienno przecikowego przyjmujac wartoci 0-1
img = color.rgb2hsv(img)

fig = plt.figure(figsize=(12, 4))

axes = []
for i in range(1, 4):
    axes.append( fig.add_subplot(1, 3, i) ) 
    
brighten_img = img.copy()
brighten_img[:,:,2] += 0.5
brighten_img[:,:,2] /= brighten_img[:,:,2].max() #znormalizujmy do 0-1

darker_img = img.copy()
darker_img[:,:,2] *= 0.5

axes[0].imshow(color.hsv2rgb(brighten_img))
axes[1].imshow(color.hsv2rgb(img))
axes[2].imshow(color.hsv2rgb(darker_img))
plt.tight_layout()
plt.show()
```
