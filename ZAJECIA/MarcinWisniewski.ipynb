```python
import os
import numpy as np

rr_data = []
for fn in sorted(os.listdir('./rr')):
    rr=open('./rr/'+fn).read()
    rr = map(float, rr.split('\n')[:-1])
    rr = np.array(rr)
    rr_data.append(rr)
```
